```python
import numpy as np
import scipy
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt 
import plotly.express as px
import datetime
%matplotlib inline
import yfinance as yf
```

```python
start = datetime.datetime (2017, 5, 1)
end = datetime.datetime (2022, 5, 1)


eur = 1/(pdr.DataReader('DEXUSEU', 'fred', start, end))
eur.tail()
```

```python
gbp = 1/(pdr.DataReader('DEXUSUK', 'fred', start, end))
gbp.tail()
```

```python
chf = (pdr.DataReader('DEXSZUS', 'fred', start, end))
chf.tail()
```

```python
df = pd.concat([eur, gbp, chf], axis = 1)
df.columns = ['USD/EUR', 'USD/GBP', 'USD/CHF']
df.tail()
```

```python
# Plot the adjusted close price
df.plot(figsize=(7, 5))

# Define the label for the title of the figure
plt.title('US Dollar to Selected European Currencies Exchange Rate', fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('US Dollar to Currency', fontsize=14)
plt.xlabel('Date', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

# Show the plot
plt.show()
```
